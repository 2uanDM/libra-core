```python
import pandas as pd
```

```python
# Specify the path to your Excel file
file_path = '../../data/MB_company_profile.xlsx'

# Read the Excel file into a DataFrame
df = pd.read_excel(file_path)
df['topic'] = df['topic'].fillna('')
```

```python
df.head(50)
```

```python
from libra.models import ModelFactory
from libra.types import ModelLabel
from dotenv import load_dotenv

load_dotenv(override=True)

from libra.models.model_factory import ModelFactory
from libra.config import ChatGPTConfig

model = ModelFactory.create(
    model_label=ModelLabel.GPT_4o,
    model_config_dict=ChatGPTConfig(stream=False, temperature=0.0).__dict__,
)
```

```python
model.run([
    {"role": "user", "content": "xin chào"}
])
```

```python
def summarize(content: str):
    response = model.run([
        {
            "role": "user", 
            "content": f"""Bạn sẽ nhận được một đoạn thông tin về ngân hàng MB trên nhiều lĩnh vực. Hãy tóm tắt đoạn văn đó bằng một đoạn ngắn gọn từ 20-50 từ:\n{content}\nChỉ trả ra đúng nội dung tóm tắt, không đưa thêm thông tin thừa thãi"""
        }
    ])
    return response.choices[0].message.content # type: ignore
```

```python
df['full_content'] = df['topic'] + "\n" + df['content']
```

```python
df[df['full_content'].isna()]
```

```python
from tqdm import tqdm

# Ensure tqdm can work with pandas
tqdm.pandas()

df['summary'] = df['full_content'].progress_apply(summarize)
```

```python
df.to_csv("../../data/mb_company_profile_summarized.csv")
```
